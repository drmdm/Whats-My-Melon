# Evaluating the Melon CNN

This notebook describes a basic CNN model used to classify three types of melon - watermelon, honeydew and canteloupe - and evaluates the performance. 

The model has already been trained using this script:
[Basic Melon CNN script](./melon_CNN.py)

I wrote a webscraper to extract 2700 images for each class here:
[Bing Webscraper](./bing_scraper.py)

Due to the size of the image folders I haven't uploaded them to github so you won't be able to run this script if you clone the repository. This just provides a template of how to run a basic CNN.  

I have included a separate prediciton notebook [CNN Prediction](./CNN_prediction.ipynb) If you want to test predictions. 

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import keras
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from keras.models import load_model
from sklearn.metrics import multilabel_confusion_matrix, classification_report
import pickle
import warnings
warnings.simplefilter('ignore')
plt.style.use('seaborn-dark')
```

Split into train, validate and test data folders

```python
datadir='./'
classes=['watermelon', 'canteloupe', 'honeydew']
train_data_dir  = datadir+'data/train'
test_data_dir   = datadir+'data/test'
```

Set up variables. Must be the same as in the training model

```python
n_train                 = 2220
n_test                  = 500
epochs                  = 50
batch_size              = 16
img_width, img_height   = 150, 150
```

Load the previously trained model and history

```python
model = load_model(datadir+'model_3_20200210T1916.h5')
with open(datadir+'history_3_20200210T1916.pickle', 'rb') as f:
    history = pickle.load(f)
```

Recreate the Test Dataset with batch_size=1 to create individual predictions.

```python
test_datagen = ImageDataGenerator(rescale=1. / 255)
test_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(img_width, img_height),
    batch_size=1,
    classes=classes,
    class_mode='categorical')

classes = list(test_generator.class_indices.keys())
n_classes=len(classes)
```

## Predict new, unseen images

```python
def predict_new_image(filepath):
    """
    Make a single prediciton on a local, previously unseenimage
    
    Parameters
    ----------
    filepath : string
        filepath to the image to predict
        
    Program Variables :
    ----------
        img_width, img_height
            As defined when training the model
        classes
            classification classes defined by:
                list(test_generator.class_indices.keys())
            or pass manually

    Returns
    -------
    prediction : list
        The class prediction probabilites for the image

    """
    #Load the image and prepare for prediction
    img = load_img(filepath, target_size=(img_width, img_height))
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x /= 255      
    images = np.vstack([x])
    
    #Make a prediction
    prediction = model.predict(images, batch_size=10)

    
    #Load the image again in full resolution for plotting
    img = load_img(filepath)
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x /= 255
    
    #Plot
    text=("What's my Melon? %s\nHow sure are you? %3.0f%%" % (classes[np.argmax(prediction[0])].capitalize(), np.max(prediction[0])*100))
    fig=plt.figure()
    plt.imshow(x[0])                           
    plt.axis('off')
    plt.title(text, loc='left')
    plt.show()
    return prediction
```

Predict all images in the folder "./data/predict".

```python
prediction=[]
filedir=datadir+'data/to_predict/'
for file in os.listdir(filedir):
    filename=file
    filepath=filedir+filename
    prediction.append(predict_new_image(filepath))
```

# Training Evaluation
Let's look at how the model performed during training.

## Training Accuracy
After a couple of trials I trained the model on 20 epochs as this seemed to be a good trade off between performance and computation. I haven't used any cross-validation due to run time. These results are for a fixed train/test split.

Let's examine how the accuracy evolved during training:

```python
fig=plt.figure()
plt.plot(history.history['acc'], c='blue')
plt.plot(history.history['val_acc'], c='red')
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

We can see the training accuracy converges a little above 60% and the test accuracy is in a similar area. A higher accuracy would be better, but we have already pointed out a number of drawbacks with the model and image collection so this isn't unexpected.

## Training Loss
Examine how the loss evolved during training

```python
fig=plt.figure()
plt.plot(history.history['loss'], c='blue')
plt.plot(history.history['val_loss'], c='red')
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()
```

Loss has reached convergence at around 0.85 for the train set and a little lower for the test set. This is a little high (0 indicates perfect predictions), but again, we know the model has flaws.

## Create predictions for the test set for evaluation

Let's create a prediction for each image in the test set so we can evaluate it. 

```python
def test_predictions(test_generator, model):
    """
    Returns the predictions for the test set
    Required to Run 'get_metrics' function

    Parameters
    ----------
    test_generator : Keras data generator
        Test Dataset
    model : keras model
        Trained Keras model

    Returns
    -------
    predictions : np.array
        Numpy array contain class prediction probabilities for all images in
        the test dataset.

    """
    
    filenames = test_generator.filenames
    nb_samples = len(filenames)
    
    #Make predictions on the test data
    predictions = model.predict_generator(test_generator, steps=nb_samples, verbose=1)
    return predictions
```

```python
predictions=test_predictions(test_generator, model)
```

## Evaluate the test dataset
This functions prints the model accuracy, classification report and confusion matrices for each class

```python
def get_metrics(predictions):  
    """
    Produce Confusion Matrices
    
    Return Model Accuracy scores


    Parameters
    ----------
    predictions : np.array
        Numpy array generated using 'test_predictions' function

    Returns
    -------
    None.

    """
    
    #Accuracy from Keras evaluator
    accuracy=model.evaluate_generator(test_generator, n_test, verbose=1)
    print('Model loss: %s' % accuracy[0])
    print('Model accuracy: %s' % accuracy[1]) 


    label_true=test_generator.labels
    label_predicted=[]
    for i in range(len(predictions)):
        label_predicted.append(np.argmax(predictions[i]))   
    
    #Print Classification Report
    cr=classification_report(label_true, label_predicted, target_names=classes)  
    print('Classification Report:\n %s' % cr)

    
    #Confusion Matrix
    matrix = multilabel_confusion_matrix(label_true, label_predicted)

    for i in range(len(classes)):
        dataframe = pd.DataFrame(matrix[i], index=[classes[i], 'not '+classes[i]], columns=[classes[i], 'not '+classes[i]])
        
        fig, ax = plt.subplots()
        sns.heatmap(dataframe, annot=True, cbar=None, cmap='Blues', fmt='g')
        #sns.heatmap(dataframe, annot=True)
        plt.title("Confusion Matrix")
        plt.ylabel("True Class")
        plt.xlabel("Predicted Class")
        ax.set_ylim([0,2])
        plt.show
```

```python
get_metrics(predictions)
```

The final loss and accuracy are 0.71 and 68% respectively.
